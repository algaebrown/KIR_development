```python
import pandas as pd
tcga_w = pd.read_csv('~/Data/liger_output/TCGA_5000_w.csv', index_col = 0, header = 0)
all_w = pd.read_csv('~/Data/liger_output/all_w.csv', index_col = 0, header = 0)
```

```python
tcga_w.head()
```

```python
t = tcga_w.idxmax()
```

```python
a = all_w.idxmax()
```

```python
kmer_df = pd.read_csv('~/Data/kmer_df', header = 0, index_col = 0)
kmer_df.set_index('kmer', inplace = True)
kmer_df = kmer_df.loc[t.index]
```

```python
kmer_df.loc[t.index, 'tcga_factor'] = t.values
```

```python
kmer_df.loc[a.index, 'all_factor'] = a.values
```

```python
kmer_df.head()
```

```python
kmer_df.loc[a.index]
```

```python
t_factor = kmer_df.groupby(by = 'tcga_factor')['gene'].value_counts().unstack().fillna(0)
```

```python
a_factor = kmer_df.groupby(by = 'all_factor')['gene'].value_counts().unstack().fillna(0)
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(t_factor, annot = True)
```

```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(a_factor, annot = True)
```
