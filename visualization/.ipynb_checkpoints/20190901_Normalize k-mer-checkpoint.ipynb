```python
# estimate coverage by 4-mer
from normalization.coverage import four_mer, four_mer_coverage
four_mers = four_mer()

# select only the qualified k-mers
from normalization.kmer_selection import select_qualified_kmer
kmer_df = pd.read_csv('~/Data/kmer_df', header = 0)
kmer_df = kmer_df.loc[kmer_df['length'] == 20]
selected_kmer = select_qualified_kmer(kmer_df['kmer'])
kmer_df = kmer_df.loc[kmer_df['kmer'].isin(selected_kmer)]

gene_kmer_count = kmer_df.groupby(by = 'gene').count()['kmer']
```

```python
gene_kmer_count.sum()
```

```python
# do it chunkwise
chunks = pd.read_csv('/cellar/users/ramarty/Data/kir/components/tcga/kir_four_random_counts.csv', index_col=0, chunksize = 1000)
chunk_no = 0

for chunk in chunks:
    # only the selected id
    chunk = chunk.loc[chunk.index.isin(selected_id)]
    
    # retrieve four mer coverages
    four_cov = four_mer_coverage(chunk, four_mers)

    # KIR reads 
    read = kir_read.loc[chunk.index, 'KIR_total_read']
    
    print('done retrieving normalization')
    # retrieve qualifies k-mer
    q_kmer = chunk[selected_kmer]
    
    # new_df
    nor_four = pd.DataFrame(index = q_kmer.index, columns = q_kmer.columns)
    nor_read = pd.DataFrame(index = q_kmer.index, columns = q_kmer.columns)

    
    # normalize by four-mer estimated coverage or KIR reads
    for row in q_kmer.index:
        nor_four.loc[row] = q_kmer.loc[row]/four_cov[row]
        nor_read.loc[row] = q_kmer.loc[row]/read[row]
    
    # save to file
    if chunk_no == 0:
        with open('/cellar/users/hsher/Data/nor_four.csv', 'w') as f:
            nor_four.to_csv(f, header = True)
    else:
        with open('/cellar/users/hsher/Data/nor_four.csv', 'a') as f:
            nor_four.to_csv(f, header = False)
    
    if chunk_no == 0:
        with open('/cellar/users/hsher/Data/nor_read.csv', 'w') as f:
            nor_read.to_csv(f, header = True)
    else:
        with open('/cellar/users/hsher/Data/nor_read.csv', 'a') as f:
            nor_read.to_csv(f, header = False)
    print(chunk_no)
    chunk_no += 1
```

```python
df = pd.read_csv('/cellar/users/hsher/Data/nor_four.csv', header = 0, index_col = 0)


from multiprocessing import Pool
def normalize(col):
    return(df[col]/denominator)

def normalize_col(gene):
    cols = kmer_df.loc[kmer_df['gene'] == gene, 'kmer']
    global denominator
    denominator = gene_kmer_count[gene]
    with Pool(16) as p:
        series = p.map(normalize, cols)
    return(pd.concat(series, axis = 1))

def normalize_by_kmer(df, output_name):
    output_df = pd.DataFrame(index = df.index)
    # normalize by count of k-mer
    for gene in gene_kmer_count.index:
        concat_df = normalize_col(gene)
        output_df = output_df.join(concat_df)
    output_df.to_csv(output_name)

d = normalize_by_kmer(df, '/cellar/users/hsher/Data/nornor_four.csv')
    
```
