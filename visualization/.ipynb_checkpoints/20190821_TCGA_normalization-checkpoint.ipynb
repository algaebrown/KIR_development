```python
# datasets
import pandas as pd
manifest = pd.read_csv('/cellar/users/ramarty/Data/kir/TCGA/manifests/whole_exome.all_samples.tsv', sep='\t') # status of TCGA file
```

```python
manifest.head()
```

```python
def valid_barcode(x):
    try: 
        x.index('TCGA')
        return True
    except:
        return False
def get_barcode(x):
    i = x.index('TCGA')
    return x[i:i+12]
def get_origin(x):
    try:
        if int(x.split('-')[3][:2]) > 9:
            return 'normal'
        else:
            return 'tumor'
    except:
        try:
            if x.split('-')[3] == 'NT' or x.split('-')[3] == 'NB':
                return 'normal'
            else:
                return 'tumor'
        except:
            return 'tumor'
manifest['valid'] = manifest.filename.apply(valid_barcode) # find which filename contains TCGA
manifest = manifest[manifest.valid] # select only those filenmae with TCGA
manifest['barcode'] = manifest.filename.apply(get_barcode) # select out the barcode (string of length 12: TCGA-QK-A6V9 from filename)
manifest['origin'] = manifest.filename.apply(get_origin) # identify who is normal who is tumor
normal = manifest[manifest.origin == 'normal']
normal = normal.drop_duplicates('barcode')
normal_samples = list(normal.id)
normal_barcodes = list(normal.barcode)
```

## Each capture kit has quite different 4-mer estimated coverage

```python
# how does capture kit affect 1. 4-mer coverage 2. anchor gene k-mer counts
import matplotlib.pyplot as plt
%matplotlib inline
each_c_kit=[]
for c_kit in capture_kits['KIT'].unique(): # for every kind of capture kit
    c_sum = df_kir_.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], four_mers].sum(axis = 1) # per patient, that kit
    each_c_kit.append(c_sum)
plt.boxplot(each_c_kit, labels = capture_kits['KIT'].unique())
plt.ylabel('No. 4-mer per sample')
plt.xticks(rotation ='vertical')
```

## And also have different KIR-specific 20-mer distribution (anchor genes only)
- anchor genes are supposed to be the same for all samples
- but they aren't. Where does that come from?

```python
# how does capture kit affect 1. 4-mer coverage 2. anchor gene k-mer counts

fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex = True, sharey = True, figsize = (3,6))
fig_count = 0
for a_gene in anchor_genes:
    each_c_kit=[]
    for c_kit in capture_kits['KIT'].unique(): # for every kind of capture kit
        c_sum = df_kir_.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], kmers[20][a_gene]].sum(axis = 1) # per patient, that kit
        each_c_kit.append(c_sum)
    axes[fig_count].boxplot(each_c_kit, labels = capture_kits['KIT'].unique())
    axes[fig_count].set_title(a_gene)
    #axes[fig_count].set_ylabel('No. 20-mer per sample')
    axes[fig_count].set_xticklabels(labels = capture_kits['KIT'].unique(), rotation ='vertical')
    fig_count += 1

fig.text(0.04, 0.5, 'No. 20-mer per sample', va='center', rotation='vertical')
plt.tight_layout()
```

## Does the difference of anchoring KIR 20-mer come from difference in coverage?
- Yes they do. The higher the 4-mer estimated coverage, the higher the KIR 20-mer count.
- it seems that some anchoring KIR are naturally hard to capture, and it is universal in each capturing kit. (see the color having different slope)

```python
# 4-mer v.s. each anchor gene coverage per capture kit


fig, axes = plt.subplots(nrows = 1, ncols = 6, sharex = True, sharey = True, figsize = (12,3))
fig_count = 0
for c_kit in capture_kits['KIT'].unique(): # for every kind of capture kit
    for a_gene in anchor_genes:
        f_sum = df_kir_.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], four_mers].sum(axis = 1) # per patient, that kit
        k_sum = df_kir_.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], kmers[20][a_gene]].sum(axis = 1) # per patient, that kit
        axes[fig_count].scatter(x = f_sum, y = k_sum, label = a_gene, alpha = 0.3)
        axes[fig_count].set_title(c_kit)
        axes[fig_count].set_ylabel('No. 20-mer per sample')
        axes[fig_count].set_xlabel('No. 4-mer per sample')
    fig_count += 1

# single shared legend
h,l = axes[5].get_legend_handles_labels()
fig.legend(h, l, bbox_to_anchor=(1.25, 0.5))
plt.tight_layout()
```

- KIR3 DL2(red) are harder to capture no matter which kit used
- KIR DL3, DP1 has higher 20-mer counts?

Why?
- maybe because some KIR has more specific 20-mers
- maybe because some KIR's specific k-mer has higher GC content which makes them preferentially amplified.

### can GC content affect k-mer count?
- calculate GC cotent for every 20-mer
- only sample 1000 patients for normalization

```python
# different anchoring kir has different capturing efficiency. And the efficiency (slope of the scatter) seems to be the same across capturing kit.
# Why? GC content of the k-mer?
def CG_content(kmer):
    return((kmer.count('C')+kmer.count('G'))/len(kmer))

# plot GC_content of each kmer v.s. that k-mer normalized by 4-mer coverage one the patient subplot by capture kit and KIR

# gc content for each k-mer
gc = pd.Series(df_kir_.columns).apply(CG_content)
gc.index = df_kir_.columns

# normalize by 4-mer count
coverage_per_sample = df_kir_.loc[pd.Series(df_kir_.index).sample(1000), four_mers].sum(axis =1)
nor_df = pd.DataFrame(index = coverage_per_sample.index, columns = all_kir_kmers) # sampled"
for col in nor_df.columns:
    nor_df[col] = df_kir_[col]/coverage_per_sample

```

```python
nor_df.shape # only 20-mer 
```

```python
for row in nor_df.index[:500]:
    plt.scatter(y = nor_df.loc[row]*100000, x = gc[nor_df.columns], alpha = 0.3, color = 'blue')
plt.ylabel('KIR k-mer/4-mer-estimated coverage*100000')
plt.xlabel('gc content')
```

- It seems that the normalized k-mer count is slightly affected by GC content. and the distribution within GC% 40-65% seems to be somewhat identical

### what will happen if we normalize by
1. 4-mer estimated coverage (per sample)
2. pick only k-mers that are of GC 40-65%
3. No. of specific k-mer per KIR anchor gene (per set of k-mer)

```python
filtered_kmers = gc.loc[all_kir_kmers].loc[gc>0.4].loc[gc<=0.65].index # select 20-mers and then filter by GC content

# count how many k-mer per genes
specific_kmers = {}
for a_gene in anchor_genes:
    specific_kmers[a_gene]=list(set(kmers[20][a_gene]).intersection(set(filtered_kmers)))

#normalize by 4-mer count
coverage_per_sample = df_kir_.loc[df_kir_.index, four_mers].sum(axis =1)
nor_nor_df = pd.DataFrame(index = df_kir_.index, columns = filtered_kmers) # sampled
for col in nor_nor_df.columns:
    nor_nor_df[col] = df_kir_[col]/coverage_per_sample
    
# normalize by number of qualified (GC = 0.4-0.65) k-mer per gene
for a_gene in specific_kmers.keys():
    nor_nor_df.loc[:, specific_kmers[a_gene]] = nor_nor_df.loc[:, specific_kmers[a_gene]]/len(specific_kmers[a_gene])
```

```python
import numpy as np
fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex = True, sharey = True, figsize = (3,7))
fig_count = 0
for a_gene in anchor_genes:
    each_c_kit=[]
    for c_kit in capture_kits['KIT'].unique(): # for every kind of capture kit
        c_sum = nor_nor_df.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], kmers[20][a_gene]].sum(axis = 1) # per patient, that kit
        each_c_kit.append(c_sum)
    axes[fig_count].boxplot(each_c_kit, labels = capture_kits['KIT'].unique())
    axes[fig_count].set_title(a_gene)
    #axes[fig_count].set_ylabel('No. 20-mer per sample')
    axes[fig_count].set_xticklabels(labels = capture_kits['KIT'].unique(), rotation ='vertical')
    fig_count += 1

fig.text(0.04, 0.5, 'No. 20-mer per sample', va='center', rotation='vertical')
fig.suptitle('normalized KIR count by 1. 4-mer estimated coverage 2. # of specific k-mer',  weight = 'extra bold', y = 1.05)
plt.tight_layout()
```

```python
# normalized KIR count by 1. 4-mer estimated coverage 2. # of specific k-mer (think of it as gene length sorta)

fig, axes = plt.subplots(nrows = 4, ncols = 1, sharex = True, sharey = True, figsize = (4,9))
fig_count = 0
for a_gene in anchor_genes:
    for c_kit in capture_kits['KIT'].unique(): # for every kind of capture kit
    
        
        k_sum = nor_nor_df.loc[capture_kits.loc[capture_kits['KIT'] == c_kit]['SHORT_ID'], kmers[20][a_gene]].sum(axis = 1)*100000 # per patient, that kit
        axes[fig_count].hist(k_sum, bins = 20, label = c_kit, alpha = 0.3, log = True)
        axes[fig_count].set_title(a_gene)
        axes[fig_count].set_xlabel('KIR specific 20-mer count normed by 4-mer coverage * 100000')
        axes[fig_count].set_ylabel('No. samples')
    fig_count += 1

# single shared legend
h,l = axes[3].get_legend_handles_labels()
fig.legend(h, l, bbox_to_anchor=(1.5, 0.5))
fig.suptitle('normalized KIR count by 1. 4-mer estimated coverage 2. # of specific k-mer',  weight = 'extra bold', y = 1.05)
plt.tight_layout()
```
