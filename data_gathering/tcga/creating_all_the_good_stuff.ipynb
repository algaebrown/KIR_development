```python
new_matrix = []
for y in hla.ix[:, :].iterrows():
    x = y[1]
    new_array = []
    # hla a
    if x[0] > x[1]:
        new_array.append(x[1])
        new_array.append(x[0])
    else:
        new_array.append(x[0])
        new_array.append(x[1])
    # hla b
    if x[2] > x[3]:
        new_array.append(x[3])
        new_array.append(x[2])
    else:
        new_array.append(x[2])
        new_array.append(x[3])
    # hla c
    if x[4] > x[5]:
        new_array.append(x[5])
        new_array.append(x[4])
    else:
        new_array.append(x[4])
        new_array.append(x[5])
    new_matrix.append(new_array)
```

```python
hla_types = pd.DataFrame(new_matrix)
hla_types.index = hla.index
```

```python
hla_types.columns = ['A1', 'A2', 'B1', 'B2', 'C1', 'C2']
```

```python
# clean up dataframe
presentation = presentation[presentation.neoantigens != '-']
presentation.neoantigens = presentation.neoantigens.astype(int)
```

```python
# create features
total_presented_peptides = presentation[['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').sum().reset_index()
total_presented_peptides.columns = ['Tumor_Sample_Barcode', 'total_presented_peptides']
```

```python
mutation_count = presentation[['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').count().reset_index()
mutation_count.columns = ['Tumor_Sample_Barcode', 'mutation_count']
```

```python
max_neoantigen = presentation[['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').max().reset_index()
max_neoantigen.columns = ['Tumor_Sample_Barcode', 'max_neoantigen']
```

```python
num_neoantigens1 = presentation[presentation.neoantigens>1][['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').count().reset_index()
num_neoantigens1.columns = ['Tumor_Sample_Barcode', 'num_neoantigens1']
num_neoantigens5 = presentation[presentation.neoantigens>5][['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').count().reset_index()
num_neoantigens5.columns = ['Tumor_Sample_Barcode', 'num_neoantigens5']
num_neoantigens10 = presentation[presentation.neoantigens>10][['Tumor_Sample_Barcode', 'neoantigens']].groupby('Tumor_Sample_Barcode').count().reset_index()
num_neoantigens10.columns = ['Tumor_Sample_Barcode', 'num_neoantigens10']
```

```python
## merge features
neoantigen_features0 = pd.merge(pd.merge(total_presented_peptides, mutation_count, on='Tumor_Sample_Barcode'),
                               pd.merge(max_neoantigen, num_neoantigens1, on='Tumor_Sample_Barcode'), on='Tumor_Sample_Barcode')
neoantigen_features = pd.merge(pd.merge(num_neoantigens5, num_neoantigens10, on='Tumor_Sample_Barcode'), neoantigen_features0, on='Tumor_Sample_Barcode')
```

```python
neoantigen_features.index = neoantigen_features.Tumor_Sample_Barcode
```

```python
neoantigen_features.head()
```

```python
# output again
merged = pd.merge(neoantigen_features, merged, left_index=True, right_index=True)
```
